# Harberman data set

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import csv

hd = pd .read_csv(r"haberman data set.csv")
print(hd.to_string())
```

```python
print(hd.shape)
print(hd.columns)
```

```python
hd.describe()
```

```python
hd.isnull().sum()
```

```python
hd["status"].value_counts()
```

```python
hd['status']=hd['status'].replace(to_replace=[1,2],value=['yes','no'])
hd
```

```python
hd.plot(kind='scatter',x='age',y='nodes');
plt.show()
```

```python
count_status = pd.value_counts(hd["status"])
count_status.plot(kind = 'bar')
plt.title("Class distribution Histogram")
plt.xlabel("Survival Status")
plt.ylabel("Frequency")
plt
```

```python
sns.set_style("whitegrid");
sns.pairplot(hd, hue="status", size=5);
plt
```

```python
sns.FacetGrid(hd, hue="status", size = 6) \
.map(sns.distplot,"age") \
.add_legend();
plt.show()
```

```python
sns.FacetGrid(hd, hue="status", size = 6) \
.map(sns.distplot,"nodes") \
.add_legend();
plt
```

```python
sns.FacetGrid(hd, hue="status", size = 6) \
.map(sns.distplot,"year") \
.add_legend();

plt
```

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 5))
for i, feature in enumerate(list(hd.columns)[:-1]):
    sns.boxplot( x='status', y=feature, data=hd, ax=axes[i])
plt.show()
```

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 5))
for i, feature in enumerate(list(hd.columns)[:-1]):
    sns.violinplot( x='status', y=feature, data=hd, ax=axes[i])
plt.show()
```

```python
plt.figure(figsize=(20,5))
for i, feature in enumerate(list(hd.columns)[:-1]):
    plt.subplot(1, 3, i+1)
    print(feature+" *********")
    counts, bin_edges = np.histogram(hd[feature], bins=10, density=True)
    print("Bin Edges: {}".format(bin_edges))
    pdf = counts/sum(counts)
    print("PDF: {}".format(pdf))
    cdf = np.cumsum(pdf)
    print("CDF: {}".format(cdf))
    plt.plot(bin_edges[1:], pdf, bin_edges[1:], cdf)
    plt.xlabel(feature)
```
